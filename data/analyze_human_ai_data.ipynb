```python
# We assume the CSV has columns like: "participant_id" (Prolific),
# "round_number" (or similar), and "accuracy" or a comparable field.

# Try to infer relevant column names:
print("Columns for reference:", list(data.columns))

# Try to intelligently guess potential names for participant/round/accuracy fields
candidate_participant_cols = [c for c in data.columns if "prolific" in c.lower() or "participant" in c.lower() or "pid" in c.lower()]
candidate_round_cols = [c for c in data.columns if "round" in c.lower()]
candidate_acc_cols = [c for c in data.columns if "acc" in c.lower() or "score" in c.lower()]

# Select first candidate or prompt
participant_col = candidate_participant_cols[0] if candidate_participant_cols else input("Enter the participant id column name: ")
round_col = candidate_round_cols[0] if candidate_round_cols else input("Enter the round number column name: ")
accuracy_col = candidate_acc_cols[0] if candidate_acc_cols else input("Enter the accuracy column name: ")

print(f"Using columns: participant = {participant_col}, round = {round_col}, accuracy = {accuracy_col}")

# Now print accuracies per round and participant
grouped = data.groupby([participant_col, round_col])[accuracy_col].mean().reset_index()

for pid in grouped[participant_col].unique():
    print(f"\nParticipant: {pid}")
    sub = grouped[grouped[participant_col] == pid]
    for _, row in sub.iterrows():
        print(f"  Round {row[round_col]}: Accuracy = {row[accuracy_col]}")
```
