### Communication Effort

```python
cols = [
    '# words',
    '# turns', 
    '# utterances', 
    'round_duration_seconds',
]
efficiency_df_annotated[cols].corr("spearman").round(2)
```

```python
cr_all = efficiency_df_annotated[cols].corr("spearman").round(2).reset_index()
cr_hh = efficiency_df_annotated[efficiency_df_annotated.condition=="Human-Human"][cols].corr("spearman").round(2).reset_index()
cr_all = cr_all[cr_all["index"].isin(["# words", "# turns"])]
cr_hh = cr_hh[cr_hh["index"].isin(["# words", "# turns"])]
```

```python
cr_all.insert(0, "Condition", "All Pairs")
cr_hh.insert(0, "Condition", "Human-Human")
combined_cr = pd.concat([cr_all, cr_hh], axis=0).reset_index(drop=True)
combined_cr.T
```

```python
print(combined_cr.to_latex(index=False, float_format="%.2f"))
```
