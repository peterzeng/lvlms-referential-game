```python
from tqdm import tqdm

extracted_descriptions = []

for _, row in tqdm(efficiency_df.iterrows(), total=len(efficiency_df)):
    pair_id = row["pair_id"]
    round_ix = row["round_ix"]
    condition = row["condition"]
    transcript = row["transcript"]
    description_dict = row["parsed_llm_extracted_object_descriptions_GPT_5"]
    for i, key in enumerate(description_dict):
        object_ix = i+1
        
        if f"object_#{object_ix}" != key:
            print(
                f"Warning: Expected object_#{object_ix} but got {key} in pair {pair_id}, round {round_ix}"
            )
            continue
        
        filename = object_filename_map[round_ix - 1]["baskets"][i]
        object_description = description_dict.get(key, None)
        extracted_descriptions.append(
            {
                "pair_id": pair_id,
                "condition": condition,
                "round_ix": round_ix,
                "transcript": transcript,
                "object_ix": object_ix,
                "filename": filename,
                "object_description": object_description,
            }
        )

extracted_descriptions_df = pd.DataFrame(extracted_descriptions)
extracted_descriptions_df.shape, extracted_descriptions_df.columns
```

```python
from __future__ import annotations

from dataclasses import dataclass
from typing import Optional, Sequence, Tuple
import re

import numpy as np
import pandas as pd


def pivot_round_diff_color_tables(
    piv: pd.DataFrame,
    *,
    index_name: str = "condition",
    round_name: str = "round_ix",
    precision: int = 1,
    # display tuning
    show_delta_for_round1: bool = False,  # usually False: round 1 has no previous
    delta_fmt: str = "paren",  # "paren" -> 62.50 (+12.50), "suffix" -> 62.50 Δ+12.50
    # colors (hex for both HTML + LaTeX)
    pos_color: str = "#1a7f37",   # green
    neg_color: str = "#d1242f",   # red
    zero_color: str = "#6e7781",  # gray
    na_color: str = "#6e7781",
    # LaTeX output tuning
    latex_float: bool = True,
    latex_star: bool = True,      # True -> table* (wide), False -> table
    latex_caption: Optional[str] = None,
    latex_label: Optional[str] = None,
    latex_escape: bool = True,
    latex_fontsize_cmd: str = r"\small",  # e.g., \small or \footnotesize
    latex_tabcolsep_pt: Optional[float] = 3.0,  # set tabcolsep; None to skip
) -> tuple[pd.DataFrame, "pd.io.formats.style.Styler", str]:
    """
    Input:
      piv: the *exact* pivot table you showed:
           - index = conditions
           - columns = MultiIndex(level0=metric header, level1=round_ix)
           - values = means (numbers)

    Output:
      - df_display: same shape as piv, but stringified with value + colored delta (round>=2)
      - styler: Jupyter-renderable (HTML) with delta colored
      - latex: pure Overleaf LaTeX (ACL-friendly booktabs + multicolumn headers)

    Notes:
      - For each metric, delta is computed *within the metric block* across rounds:
            delta[r] = value[r] - value[r-1]
      - Each cell shows: "value (±delta)" with delta colored green/red/gray.
    """
    if not isinstance(piv.columns, pd.MultiIndex) or piv.columns.nlevels < 2:
        raise ValueError("piv must have MultiIndex columns with (metric, round_ix).")

    def _hex_to_rgb(hex_color: str) -> tuple[int, int, int]:
        s = hex_color.strip()
        if s.startswith("#"):
            s = s[1:]
        if len(s) == 3:
            s = "".join([ch * 2 for ch in s])
        if not re.fullmatch(r"[0-9a-fA-F]{6}", s):
            s = "000000"
        return int(s[0:2], 16), int(s[2:4], 16), int(s[4:6], 16)

    def _latex_escape(s: str) -> str:
        if not latex_escape:
            return s
        repl = {
            "\\": r"\textbackslash{}",
            "&": r"\&",
            "%": r"\%",
            "$": r"\$",
            "#": r"\#",
            "_": r"\_",
            "{": r"\{",
            "}": r"\}",
            "~": r"\textasciitilde{}",
            "^": r"\textasciicircum{}",
        }
        return "".join(repl.get(ch, ch) for ch in s)

    def _is_num(x) -> bool:
        try:
            return np.isfinite(float(x))
        except Exception:
            return False

    # ---- normalize & sort columns (metric, round) ----
    cols = piv.columns
    metric_level = 0
    round_level = 1

    metrics = list(pd.unique(cols.get_level_values(metric_level)))
    # try numeric sorting for round_ix
    rounds_raw = cols.get_level_values(round_level)
    try:
        rounds = sorted(pd.unique(rounds_raw), key=lambda z: float(z))
    except Exception:
        rounds = list(pd.unique(rounds_raw))

    # reindex columns in metric-major, round-minor order
    ordered_cols = []
    for m in metrics:
        for r in rounds:
            if (m, r) in cols:
                ordered_cols.append((m, r))
    piv = piv.copy().reindex(columns=pd.MultiIndex.from_tuples(ordered_cols, names=cols.names))

    # ---- compute deltas ----
    delta = pd.DataFrame(index=piv.index, columns=piv.columns, dtype=float)
    for m in metrics:
        for j, r in enumerate(rounds):
            if (m, r) not in piv.columns:
                continue
            if j == 0:
                delta[(m, r)] = np.nan
            else:
                prev_r = rounds[j - 1]
                if (m, prev_r) in piv.columns:
                    v_cur = pd.to_numeric(piv[(m, r)], errors="coerce")
                    v_pre = pd.to_numeric(piv[(m, prev_r)], errors="coerce")
                    delta[(m, r)] = v_cur - v_pre
                else:
                    delta[(m, r)] = np.nan

    # ---- build HTML display strings (Jupyter) ----
    def _delta_color_hex(d: float) -> str:
        if not np.isfinite(d):
            return na_color
        if d > 0:
            return pos_color
        if d < 0:
            return neg_color
        return zero_color

    def _fmt_num(x: float) -> str:
        if not np.isfinite(x):
            return ""
        return f"{x:.{precision}f}"

    def _fmt_delta(d: float) -> str:
        if not np.isfinite(d):
            return ""
        sign = "+" if d > 0 else ""
        return f"{sign}{d:.{precision}f}"

    df_display_html = pd.DataFrame(index=piv.index, columns=piv.columns, dtype=object)
    for m in metrics:
        for j, r in enumerate(rounds):
            if (m, r) not in piv.columns:
                continue
            for idx in piv.index:
                v = piv.loc[idx, (m, r)]
                v_num = float(v) if _is_num(v) else np.nan
                v_str = _fmt_num(v_num) if np.isfinite(v_num) else ""

                d = delta.loc[idx, (m, r)]
                has_prev = (j > 0)
                show_delta = has_prev or show_delta_for_round1

                if not show_delta or (not np.isfinite(d)):
                    # round 1 (or missing) -> just value
                    df_display_html.loc[idx, (m, r)] = v_str
                    continue

                d_str = _fmt_delta(float(d))
                col = _delta_color_hex(float(d))
                if delta_fmt == "suffix":
                    df_display_html.loc[idx, (m, r)] = (
                        f"{v_str} <span style='color:{col};'>Δ{d_str}</span>"
                    )
                else:  # "paren"
                    df_display_html.loc[idx, (m, r)] = (
                        f"{v_str} <span style='color:{col};'>({d_str})</span>"
                    )

    # Make a Styler that does NOT escape HTML so <span> works
    styler = (
        df_display_html.style
        .format(escape="html")  # key: allow <span> coloring
        .set_table_styles([
            {"selector": "th", "props": [("text-align", "left")]},
            {"selector": "td", "props": [("text-align", "right")]},
        ])
    )

    # ---- build LaTeX ----
    pos_rgb = _hex_to_rgb(pos_color)
    neg_rgb = _hex_to_rgb(neg_color)
    zero_rgb = _hex_to_rgb(zero_color)
    na_rgb = _hex_to_rgb(na_color)

    def _latex_delta_wrap(d: float, inner: str) -> str:
        # inner is already escaped (except \textcolor wrapper)
        if not np.isfinite(d):
            # return r"\textcolor{naColor}{" + inner + "}"
            return inner
        if d > 0:
            return r"{\tiny\textcolor{posColor}{" + inner + "}}"
        if d < 0:
            return r"{\tiny\textcolor{negColor}{" + inner + "}}"
        # d == 0 → no coloring
        return inner

    def _latex_cell(v: float, d: float, show_delta: bool) -> str:
        v_str = _fmt_num(v) if np.isfinite(v) else ""
        if not show_delta or (not np.isfinite(d)):
            return _latex_escape(v_str)

        d_str = _fmt_delta(d)
        if delta_fmt == "suffix":
            inner = r"Δ" + _latex_escape(d_str)
            return _latex_escape(v_str) + " " + _latex_delta_wrap(d, inner)
        else:
            inner = "(" + _latex_escape(d_str) + ")"
            return _latex_escape(v_str) + " " + _latex_delta_wrap(d, inner)

    # Column spec: l + (num_rounds * num_metrics) times c
    n_rounds = sum(1 for r in rounds if any((m, r) in piv.columns for m in metrics))
    n_cols = len(ordered_cols)
    col_spec = "l" + "c" * n_cols

    # header rows: metric row (multicolumn), round row (1..4 under each metric)
    # build per-metric spans (some metrics might not have all rounds present)
    metric_spans: list[tuple[str, list]] = []
    for m in metrics:
        m_cols = [(m, r) for r in rounds if (m, r) in piv.columns]
        if m_cols:
            metric_spans.append((str(m), m_cols))

    lines: list[str] = []
    if latex_float:
        env = "table*" if latex_star else "table"
        lines.append(rf"\begin{{{env}}}[t]")
        lines.append(r"\centering")
    if latex_fontsize_cmd:
        lines.append(latex_fontsize_cmd)
    if latex_tabcolsep_pt is not None:
        lines.append(rf"\setlength{{\tabcolsep}}{{{latex_tabcolsep_pt}pt}}")

    # color defs (xcolor)
    lines.append(r"\definecolor{posColor}{RGB}{" + f"{pos_rgb[0]},{pos_rgb[1]},{pos_rgb[2]}" + r"}")
    lines.append(r"\definecolor{negColor}{RGB}{" + f"{neg_rgb[0]},{neg_rgb[1]},{neg_rgb[2]}" + r"}")
    lines.append(r"\definecolor{zeroColor}{RGB}{" + f"{zero_rgb[0]},{zero_rgb[1]},{zero_rgb[2]}" + r"}")
    lines.append(r"\definecolor{naColor}{RGB}{" + f"{na_rgb[0]},{na_rgb[1]},{na_rgb[2]}" + r"}")

    lines.append(r"\begin{tabular}{" + col_spec + r"}")
    lines.append(r"\toprule")

    # metric header
    metric_header_cells = [_latex_escape(index_name)]
    cmidrules: list[str] = []
    col_cursor = 2  # LaTeX columns are 1-indexed; col 1 is the row label
    for m, m_cols in metric_spans:
        span = len(m_cols)
        metric_header_cells.append(r"\multicolumn{" + str(span) + r"}{c}{" + _latex_escape(m) + r"}")
        cmidrules.append(r"\cmidrule(lr){" + f"{col_cursor}-{col_cursor+span-1}" + r"}")
        col_cursor += span
    lines.append(" & ".join(metric_header_cells) + r" \\")
    lines.append("".join(cmidrules))

    # round header
    round_header_cells = [""]  # empty under the row label
    for m, m_cols in metric_spans:
        for (_mm, r) in m_cols:
            round_header_cells.append(_latex_escape(str(r)))
    lines.append(" & ".join(round_header_cells) + r" \\")
    lines.append(r"\midrule")

    # body
    for cond in piv.index:
        row = [_latex_escape(str(cond))]
        for (m, r) in ordered_cols:
            v = piv.loc[cond, (m, r)]
            v_num = float(v) if _is_num(v) else np.nan
            d = delta.loc[cond, (m, r)]
            j = rounds.index(r) if r in rounds else 0
            show_delta = (j > 0) or show_delta_for_round1
            row.append(_latex_cell(v_num, float(d) if np.isfinite(d) else np.nan, show_delta))
        lines.append(" & ".join(row) + r" \\")
    lines.append(r"\bottomrule")
    lines.append(r"\end{tabular}")

    if latex_caption:
        lines.append(r"\caption{" + _latex_escape(latex_caption) + r"}")
    if latex_label:
        lines.append(r"\label{" + _latex_escape(latex_label) + r"}")
    if latex_float:
        env = "table*" if latex_star else "table"
        lines.append(rf"\end{{{env}}}")

    latex = "\n".join(lines)

    # A plain-string display DF (helpful for quick inspection / exporting)
    # For that, we’ll strip HTML but keep "(+x.xx)".
    df_display_plain = pd.DataFrame(index=piv.index, columns=piv.columns, dtype=object)
    for m in metrics:
        for j, r in enumerate(rounds):
            if (m, r) not in piv.columns:
                continue
            for idx in piv.index:
                v = piv.loc[idx, (m, r)]
                v_num = float(v) if _is_num(v) else np.nan
                v_str = _fmt_num(v_num) if np.isfinite(v_num) else ""
                d = delta.loc[idx, (m, r)]
                show_delta = (j > 0) or show_delta_for_round1
                if not show_delta or (not np.isfinite(d)):
                    df_display_plain.loc[idx, (m, r)] = v_str
                else:
                    d_str = _fmt_delta(float(d))
                    if delta_fmt == "suffix":
                        df_display_plain.loc[idx, (m, r)] = f"{v_str} Δ{d_str}"
                    else:
                        df_display_plain.loc[idx, (m, r)] = f"{v_str} ({d_str})"

    return df_display_plain, styler, latex
```
